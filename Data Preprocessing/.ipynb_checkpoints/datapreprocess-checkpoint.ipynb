## We have use sklearn to add sum values to null field for that we use Imputer

##  first we choose the column for which column we want fill the numeric values and string values. 
##     ----- For that we use iloc function basically it works on row and col wise operation
## syntax :
###              iloc[from : end-1,from:end-1]


```python
dataselection=data.iloc[:,0:5]
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent') # here the missing values = np.nan (it will cj=heck for nan field and fill with most reapted values(we can do operation like mean,median ,std,etc))
```

```python
dataselection  = imputer.fit_transform(dataselection)
```

## For category of strings included we used most frequent stratagy other than mean ,mode ,etc..........................................
## for numeric null we use mode,mean, most_frequent,etc.....
